# Linear Regression

```python
from sklearn import linear_model, datasets
from pprint import pprint
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import SGDRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
reg = linear_model.LinearRegression()
```

```python
data = datasets.load_diabetes()
```

```python
data.keys()
```

```python
train = data['data']
label = data['target']
```

```python
for x,y in zip(train,label):
    print(x,y)
```

```python
reg.fit(train,label)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
predicted_vals = reg.predict(train)
```

```python
mean_squared_error(label, predicted_vals)
```

```python
# plt.plot(train, predicted_vals, "r-")
plt.plot(train, label, "b.")
plt.show()
```

#### This is the error without training the model
#### Now we will train the model using Stochastic Gradient Descent Regressor

```python
reg = make_pipeline(StandardScaler(),
                     SGDRegressor(loss='squared_loss', alpha=0.005, eta0=0.0009, tol=1e-5, 
                                  fit_intercept=True, max_iter=5000, shuffle=True, verbose=1))
```

```python
reg.fit(train,label)
```

```python
sgd_predicted_vals = reg.predict(train)
```

```python
mean_squared_error(label, sgd_predicted_vals)
```

# Normal Equation

```python
import numpy as np
```

```python
theta_best = np.linalg.pinv(train.T.dot(train)).dot(train.T).dot(label)
```

```python
theta_best
```

```python
y_predict = train.dot(theta_best) # Instead of training dataset, we'll use testing dataset
```

```python
mean_squared_error(label, y_predict)
```
